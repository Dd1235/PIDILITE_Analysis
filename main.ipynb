# Calculation of Different Ratios related to Balance Sheet, Income Statement and Cash Flow Statement

```python
# Three major financial statements: Balance Sheet, Income Statement, Cash Flow
balance_sheet = pidilite.balance_sheet
income_statement = pidilite.financials
cash_flow = pidilite.cashflow
```

```python
# Transpose the DataFrames to have years as rows
balance_sheet_df = balance_sheet.T # type: ignore
income_statement_df = income_statement.T
cash_flow_df = cash_flow.T
```

Since this is a rudimentary EDA, I will only keep the columns relevant to the ratios I will calculate

```python
balance_sheet_df = balance_sheet_df[['Cash And Cash Equivalents', 'Gross Accounts Receivable', 'Other Short Term Investments', 
                                     'Current Liabilities', 'Total Debt', 'Stockholders Equity', 'Total Assets', 
                                     'Total Liabilities Net Minority Interest']]
income_statement_df = income_statement_df[['Total Revenue', 'Cost Of Revenue', 'Net Income', 'Operating Income']]
cash_flow_df = cash_flow_df[['Operating Cash Flow', 'Capital Expenditure']]
```

```python
# Balance Sheet Ratios
current_assets = balance_sheet_df[['Cash And Cash Equivalents', 'Gross Accounts Receivable', 'Other Short Term Investments']].sum(axis=1)

current_liabilities = balance_sheet_df['Current Liabilities']

total_debt = balance_sheet_df['Total Debt']

stockholders_equity = balance_sheet_df['Stockholders Equity']

total_assets = balance_sheet_df['Total Assets']

total_liabilities = balance_sheet_df['Total Liabilities Net Minority Interest']
```

```python
# Creating a new DataFrame with only necessary columns and calculated ratios
analysis_df = pd.DataFrame(index=balance_sheet_df.index)
analysis_df['Current Assets'] = current_assets
analysis_df['Current Liabilities'] = current_liabilities
analysis_df['Total Debt'] = total_debt
analysis_df['Stockholders Equity'] = stockholders_equity
analysis_df['Total Assets'] = total_assets
analysis_df['Total Liabilities'] = total_liabilities
```

```python
# calculated from balance sheet
# Liquidity ratio
analysis_df['Current Ratio'] = current_assets / current_liabilities
# Leverage ratios 
analysis_df['Debt to Equity Ratio'] = total_debt / stockholders_equity
analysis_df['Equity Multiplier'] = total_assets / stockholders_equity
# Solvency ratio
analysis_df['Debt to Assets Ratio'] = total_debt / total_assets
```

```python
total_revenue = income_statement_df['Total Revenue']
cost_of_revenue = income_statement_df['Cost Of Revenue']
net_income = income_statement_df['Net Income']
operating_income = income_statement_df['Operating Income']
```

```python
# calculated from income sheet
# Efficiency ratios
analysis_df['Gross Profit Margin'] = (total_revenue - cost_of_revenue) / total_revenue
analysis_df['Operating Profit Margin'] = operating_income / total_revenue
```

```python
# ratio from income sheet and balance sheet
# asset turnover ratio
analysis_df['Asset Turnover Ratio'] = total_revenue / total_assets
```

```python
# Add new ratios
operating_cash_flow = cash_flow_df['Operating Cash Flow']

analysis_df['Cash Flow to Net Income Ratio'] = operating_cash_flow / net_income
analysis_df['Operating Cash Flow Ratio'] = operating_cash_flow / current_liabilities
```

```python
# Calculate average of each ratio and add it as a new row
average_ratios = analysis_df.mean()
average_ratios.name = 'Average'
analysis_df = analysis_df.append(average_ratios)
```

```python
print("\nFinal DataFrame with Ratios:\n", analysis_df)
```

```python
# Impute missing values in the analysis_df for visualization purpose
analysis_df.fillna(method='ffill', inplace=True)  # Forward fill
analysis_df.fillna(method='bfill', inplace=True)  # Backward fill if forward fill doesn't work
```

```python
 
# Select only the ratios and margins for plotting
ratios_and_margins_df = analysis_df[['Current Ratio', 'Debt to Equity Ratio', 'Debt to Assets Ratio', 
                                     'Gross Profit Margin', 'Operating Profit Margin', 'Asset Turnover Ratio', 'Cash Flow to Net Income Ratio', 'Operating Cash Flow Ratio']]
ratios_and_margins_df = ratios_and_margins_df.drop('Average')

# Plotting the ratios and margins over the years
plt.figure(figsize=(14, 8))
for column in ratios_and_margins_df.columns:
    plt.plot(ratios_and_margins_df.index, ratios_and_margins_df[column], marker='o', label=column)

plt.title('Financial Ratios and Margins Over Time')
plt.xlabel('Year')
plt.ylabel('Ratio/Margin Value')
plt.legend()
plt.grid(True)
plt.show()
```

# Analysis

1.	Current Ratio:

	A current ratio above 1 indicates that the company has more current assets than current liabilities, suggesting good short-term financial health and liquidity.
2.  Debt to Equity Ratio:
	A low debt to equity ratio is a sign of low financial risk, showing that the company primarily uses equity financing rather than debt.
3.	Debt to Assets Ratio:
	A low debt to assets ratio indicates that a small portion of the company’s assets is financed through debt, further emphasizing low financial risk.
4.	Gross Profit Margin:
	Consistent gross profit margins indicate stable cost control in relation to revenue, suggesting that the company effectively manages its production costs.
5.	Operating Profit Margin:
	While the increasing trend is positive, the overall low operating profit margin indicates that the company has relatively high operating expenses compared to its revenue. This is not uncommon in blue-chip companies.
6.	Asset Turnover Ratio:
	A stable asset turnover ratio suggests that the company maintains consistent efficiency in using its assets to generate revenue.
7.	Cash Flow to Net Income Ratio:
	A high cash flow to net income ratio suggests that the company generates substantial cash from its core activities, which is a positive indicator of financial health.
8.	Operating Cash Flow Ratio:
	An increasing operating cash flow ratio indicates that the company is generating sufficient cash to cover its short-term liabilities from its operating activities, demonstrating strong liquidity and operational efficiency.

# Basic Stock Price Analysis

```python
stock_data = pidilite.history(period="5y")

# Calculate key metrics
stock_data['Pct Change'] = stock_data['Close'].pct_change()
stock_data['20 Day MA'] = stock_data['Close'].rolling(window=20).mean()
stock_data['50 Day MA'] = stock_data['Close'].rolling(window=50).mean()
stock_data['200 Day MA'] = stock_data['Close'].rolling(window=200).mean()
stock_data['Volatility'] = stock_data['Close'].rolling(window=20).std()

# Drop NaN values
stock_data.dropna(inplace=True)


# Plot the stock price and moving averages
plt.figure(figsize=(14, 8))
plt.plot(stock_data.index, stock_data['Close'], label='Close Price')
plt.plot(stock_data.index, stock_data['20 Day MA'], label='20 Day MA')
plt.plot(stock_data.index, stock_data['50 Day MA'], label='50 Day MA')
plt.plot(stock_data.index, stock_data['200 Day MA'], label='200 Day MA')
plt.title('Pidilite Industries Ltd. Stock Price and Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price (INR)')
plt.legend()
plt.grid(True)
plt.show()

```

Pidilite Industries Ltd.’s stock demonstrates a strong overall upward trend with periodic corrections, typical of a growing company in a bullish market.
